```python
import pandas as pd
train = pd.read_csv('..\\Coursework\\Dataset\\train.csv', usecols=["payprice","click"])
validation = pd.read_csv('..\\Coursework\\Dataset\\validation.csv', usecols=["payprice","click"])
```

```python
train["payprice"].mean()
```

```python
validation["payprice"].mean()
```

```python
train_payprice = train["payprice"].sample(303925)
```

```python
train_click = train['click'].loc[train_payprice.index]
```

```python
validation_payprice = validation["payprice"].sample(303925)
```

```python
validation_click = validation['click'].loc[validation_payprice.index]
```

```python
array_bidprice = []
array_clicks = []
array_win = []
for i in range(70,80):
    sum = 6250 * 1000
    clicks = 0
    win = 0
    bid = 0
    array_bidprice.append(i)
    for pay,click in zip(validation_payprice, validation_click):
        sum_sub = sum - pay
        bidprice = 78.15
#         bidprice = 71
#         bidprice = 146
#         bidprice = 181
#         bidprice = 74.575
#         bidprice = 68.0
#         bidprice = 78.23
        if sum_sub >= 0:
            bid += 1
        if bidprice-pay > 0 and sum_sub >= 0:
            sum = sum_sub
            clicks += click
            win += 1
    array_clicks.append(clicks)
    print(i, bid, clicks, win, sum)
print(array_bidprice, array_clicks)
```
